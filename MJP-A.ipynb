# A-PUNJAB STATE ANALYTICS FROM 15 JUNE 2020 TILL NOW

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from datetime import datetime as dt
punjab= pd.read_csv("complete.csv")
punjab
```

```python
punjab.dtypes # to check the type of DATE column
```

```python
plt.plot(punjab['Date'], punjab['Total Confirmed cases'], label = "Total Confirmed cases", color="blue")
#plt.plot(punjab['Date'], punjab['Death'], label = "Total Deaths")
plt.plot(punjab['Date'], punjab['Cured/Discharged/Migrated'], label = "Total Cured", color="orange")
plt.ylabel('CASES')

plt.title('GROWTH OF COVID 19 CASES IN Punjab till now')
plt.xticks(punjab['Date'],rotation=90)
plt.legend(loc='best')
plt.yticks(np.arange(0,8000,500))
plt.plot(punjab['Date'], punjab['Total Confirmed cases'])
for a,b in zip(punjab['Date'], punjab['Total Confirmed cases']): 
    plt.text(a, b, str(b),rotation=90)
plt.plot(punjab['Date'], punjab['Cured/Discharged/Migrated'])
for a,b in zip(punjab['Date'],punjab['Cured/Discharged/Migrated']): 
    plt.text(a, b, str(b),rotation=90)
plt.show()
```

```python
plt.plot(punjab['Date'], punjab['New cases'], label = "New cases", color="blue")
#plt.plot(punjab['Date'], punjab['Death'], label = "Total Deaths")
plt.plot(punjab['Date'], punjab['New recovered'], label = "New recovered", color="orange")
plt.ylabel('CASES')

plt.title('DAY WISE COVID 19 CASES IN Punjab till now')
plt.xticks(punjab['Date'],rotation=90)
plt.legend(loc='best')
plt.yticks(np.arange(0,8000,500))
plt.plot(punjab['Date'], punjab['New cases'])
for a,b in zip(punjab['Date'], punjab['New cases']): 
    plt.text(a, b, str(b),rotation=90)
plt.plot(punjab['Date'], punjab['New recovered'])
for a,b in zip(punjab['Date'],punjab['New recovered']): 
    plt.text(a, b, str(b),rotation=90)
plt.show()
```

```python

plt.plot(punjab['Date'], punjab['Death'], label = "Total Deaths", color="blue")

plt.ylabel('CASES')

plt.title('GROWTH OF COVID 19 DEATH CASES IN Punjab till now')
plt.xticks(punjab['Date'],rotation=90)
plt.legend(loc='best')
#plt.rcParams['figure.figsize']=(1400,100)
#plt.xticks(np.arange(0,10,1))
plt.yticks(np.arange(0,8000,500))
plt.plot(punjab['Date'], punjab['Death'])
for a,b in zip(punjab['Date'], punjab['Death']): 
    plt.text(a, b, str(b),rotation=90)

plt.show()
```

```python
plt.plot(punjab['Date'], punjab['New deaths'], label = "New deaths", color="blue")

plt.ylabel('CASES')

plt.title('DAILY COVID 19 DEATH CASES IN Punjab till now')
plt.xticks(punjab['Date'],rotation=90)
plt.legend(loc='best')
#plt.rcParams['figure.figsize']=(1400,100)
#plt.xticks(np.arange(0,10,1))
plt.yticks(np.arange(0,8000,500))
plt.plot(punjab['Date'], punjab['New deaths'])
for a,b in zip(punjab['Date'], punjab['New deaths']): 
    plt.text(a, b, str(b),rotation=90)

plt.show()
```
